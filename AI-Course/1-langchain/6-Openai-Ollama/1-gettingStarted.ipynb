```python
from langchain_openai import ChatOpenAI

llm=ChatOpenAI(model="gpt-4o")
llm
```

```python
## Input and get response from LLM

result = llm.invoke("What is generative AI ?")
result
# So whatever we are hiting will be seen in langsmith what queries we have hitted
```

```python
## ChatPrompt template 
## You will tell your llm model what is the role of the model and what is the input you are providing to the model.

from langchain_core.prompts import ChatPromptTemplate

prompt=ChatPromptTemplate.from_messages(
    [
        ("system","You are a expert AI engineer, provide answers based on the question "),
        ("user","{input}")
    ]
)
prompt
```

```python
#Chain the prompt and llm together to get the response from the llm model
chain=prompt| llm
response=chain.invoke({"input":"Can you tell me about Langsmith?"})
response
```

```python
type(response)
```

```python
## String output parser 

from langchain_core.output_parsers import StrOutputParser
output_parser=StrOutputParser()
chain=prompt|llm|output_parser
response=chain.invoke({"input":"Can you tell me about Langsmith?"})
response
```
