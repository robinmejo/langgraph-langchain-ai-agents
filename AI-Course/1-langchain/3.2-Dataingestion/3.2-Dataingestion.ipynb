Data Ingestion Using Langchain

```python
## Text loader

from langchain_community.document_loaders import TextLoader

loader =TextLoader("speech.txt")
loader
```

```python
text_documents=loader.load()
text_documents
```

```python
## Reading a pdf file

from langchain_community.document_loaders import PyPDFLoader

pdf_data=PyPDFLoader("attention.pdf")
pdf_documents=pdf_data.load()
pdf_documents
```

```python
print(type(pdf_documents))
print(type(pdf_documents[0]))
```

```python
## Web Based loader
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(
    web_paths=("https://lilianweng.github.io/posts/2023-06-23-agent/",),
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            class_=("post-title", "post-content", "post-header")
        )
    )
)

docs = loader.load()

print(docs[0].page_content[:200])
```

```python
from langchain_community.document_loaders import WebBaseLoader
import bs4

loader = WebBaseLoader(
    web_paths=("https://www.geeksforgeeks.org/what-is-artificial-intelligence/",),
    bs_kwargs=dict(
        parse_only=bs4.SoupStrainer(
            "div",
            class_="entry-content"
        )
    )
)

web_documents = loader.load()

web_documents
```

```python
from langchain_community.document_loaders import ArxivLoader

docs = ArxivLoader(
    query="1605.08386",#paper number this is
    load_max_docs=2
).load()

print(len(docs))
print(docs[0].page_content[:200])
```

```python
from langchain_community.document_loaders import WikipediaLoader

docs = WikipediaLoader(
    query="Artificial Intelligence",
    load_max_docs=2
).load()

print(len(docs))
print(docs[0].page_content[:250])
```
