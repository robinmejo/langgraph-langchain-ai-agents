#### Facebook AI Similarity Search (FAISS)

Facebook AI Similarity Search (**FAISS**) is a library for efficient similarity search and clustering of dense vectors. It contains algorithms that search in sets of vectors of any size, including ones that may not fit entirely in RAM. It also provides supporting code for evaluation and parameter tuning.

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter

## install faiss-cpu version for local machine, if you are using GPU then install faiss-gpu version

loader = TextLoader("speech.txt")
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs=text_splitter.split_documents(documents)
print(docs[0].page_content)
```

```python
embeddings = OllamaEmbeddings(model="nomic-embed-text")

db = FAISS.from_documents(docs, embeddings)
db
```

```python
## Querying the vector store

query = "What Cloud computing enables?"
docs = db.similarity_search(query)
docs[0].page_content
```

📚 As a Retriever

🔍 A Retriever is an interface that searches a vector database and returns the most relevant documents for a given query.

✅ It converts a Vector Store (like FAISS or Chroma) into a Retriever, making it easy to use with LangChain chains and RAG applications.

✨ Why use a Retriever?
🔎 Retrieves the most relevant documents.
⚡ Simplifies searching in vector databases.
🔗 Integrates easily with LangChain components.
🤖 Commonly used in Retrieval-Augmented Generation (RAG)

💻 Example
retriever = db.as_retriever()

docs = retriever.invoke("Why is continuous learning important?")

```python
retriever = db.as_retriever()
docs = retriever.invoke(query)
docs[0].page_content
```

📚 Similarity Search with Score

🔍 similarity_search_with_score() retrieves the most relevant documents along with a similarity (distance) score.

📏 The score represents the L2 (Euclidean) distance between the query embedding and the document embedding.

✨ Why use it?
📄 Returns the matching documents.
📊 Also returns a similarity (distance) score.
✅ Helps evaluate how closely a document matches the query.
📝 Important
🟢 Lower score = Better match (more similar)
🔴 Higher score = Less similar

💻 Example
results = db.similarity_search_with_score(
    "Why is continuous learning important?"
)

for doc, score in results:
    print(score)
    print(doc.page_content)

```python
results = db.similarity_search_with_score(query)

for doc, score in results:
    print(score)
    print(doc.page_content)
```
