📚 Chroma

🗄️ Chroma is an open-source vector database used to store, manage, and search embeddings efficiently. It is widely used in RAG (Retrieval-Augmented Generation) applications.

✨ Why use Chroma?
- 📄 Stores document embeddings.
- 🔍 Performs fast semantic similarity searches.
- 💾 Supports persistent storage (save and reload data).
- 🔗 Integrates seamlessly with LangChain.
- 🤖 Commonly used in AI chatbots and RAG systems.

🚀 Features
- ⚡ Fast vector search
- 💽 Persistent database support
- 📚 Metadata filtering
- 🔗 Easy LangChain integration

```python
## Building a sample vectordb

from langchain_chroma import Chroma
from langchain_community .document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader("speech.txt")
data=loader.load()
data
```

```python
# Splitting the document into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
chunks = text_splitter.split_documents(data)
```

```python
#Vectorization of the chunks along with the embeddings

embeddings = OllamaEmbeddings(model="nomic-embed-text")
vectorstore = Chroma.from_documents(documents=chunks,embedding=embeddings)
vectorstore
```

```python
## Querying the vector store

Query = "What is the main topic of the speech?"
docs = vectorstore.similarity_search(Query)
docs[0].page_content
```

```python
## Saving to the disk
vectordb = Chroma.from_documents(documents=chunks, embedding=embeddings, persist_directory="./chroma_db")
```

```python
#Loading from the disk
db2 = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
docs2 = db2.similarity_search(Query)
print(docs2[0].page_content)
```

```python
## Querying the vector store
docs=db2.similarity_search_with_score("What is the main topic of the speech?")
docs
```

```python
# Retriever option
retriever = db2.as_retriever(search_type="similarity", search_kwargs={"k": 2})
retriever.invoke(Query)
```

🧩 Breakdown
- 🗄️ db2 → Your vector database (FAISS/Chroma).
- 🔄 as_retriever() → Converts the vector database into a Retriever.
- 🔍 search_type="similarity" → Uses similarity search to find the closest matching documents.
- 📄 search_kwargs={"k": 2} → Returns the top 2 most relevant documents.
