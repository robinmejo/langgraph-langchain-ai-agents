```python
from langchain_ollama import OllamaEmbeddings

embeddings = OllamaEmbeddings(model="nomic-embed-text")

r1 = embeddings.embed_documents([
    "The sky is blue because of the way sunlight interacts with Earth's atmosphere.",
    "The sky appears blue to human eyes due to the scattering of sunlight by the gases and particles in the atmosphere."
])

print(len(r1))
print(len(r1[0]))
```

```python
vector = embeddings.embed_query("Why is the sky blue?")

print(type(vector))
print(len(vector))
print(vector[:5])  # Print the first 5 values

from langchain_ollama import ChatOllama

llm = ChatOllama(model="gemma2:2b")

response = llm.invoke(
    "Answer using this context:\n"
    "The sky is blue because of Rayleigh scattering.\n\n"
    "Question: Why is the sky blue?"
)

print(response.content)
```
