```python
from langchain_openai import OpenAIEmbeddings

openai_embeddings = OpenAIEmbeddings(model="text-embedding-3-large")  # You can also use "text-embedding-3-small" for smaller embeddings
openai_embeddings
```

```python
text="This is a test text to generate embeddings using OpenAI's embedding model."
embedding = openai_embeddings.embed_query(text)
embedding
```

```python
len(embedding)  # This will give you the length of the embedding vector
```

```python
embedding_1024=openai_embeddings.embed_query(text, model="text-embedding-3-small", dimensions=1024)  # This will generate a 1024-dimensional embedding,
# or u can use specify the model as "text-embedding-3-small" to get a 1024-dimensional embedding
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader =TextLoader("speech.txt")
docs=loader.load()
docs

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=20
)

final_documents = text_splitter.split_documents(docs)
final_documents
```

```python
## Vector Embedding and Vector Store DB

from langchain_community.vectorstores import Chroma

db=Chroma.from_documents(final_documents, openai_embeddings)  # This will create a vector store 
# from the documents using the specified embeddings
db
```

```python
query="Welcome to the annual technology conference."
retrieved_results = db.similarity_search(query, k=2)  # This will return the top 2 most similar documents to the query
print(retrieved_results)
```
