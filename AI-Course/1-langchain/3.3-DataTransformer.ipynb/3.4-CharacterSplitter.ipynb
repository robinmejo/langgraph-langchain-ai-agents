### CharacterTextSplitter

- The simplest text splitter in LangChain.
- Splits text based on a specified character separator.
- Chunk size is measured by the number of characters.

**Key Points**
- Text is split using a single character separator.
- Chunk size is measured by the number of characters.

**Parameters**
- `separator` – Character used to split the text (e.g., `"\n"` or `" "`).
- `chunk_size` – Maximum number of characters in each chunk.
- `chunk_overlap` – Number of overlapping characters between consecutive chunks.

```python
from langchain_community.document_loaders import TextLoader

loader=TextLoader("speech.txt")
docs=loader.load()
docs
```

```python
from langchain_text_splitters import CharacterTextSplitter
text_splitter=CharacterTextSplitter(separator="\n",chunk_size=100, chunk_overlap=0)
texts=text_splitter.split_documents(docs)
texts
```

```python
with open("speech.txt", "r") as file:
    content = file.read()

text_splitter = CharacterTextSplitter(separator="\n", chunk_size=100, chunk_overlap=0)
chunks = text_splitter.split_text(content)   
print(chunks[0])
print(chunks[1])
```
