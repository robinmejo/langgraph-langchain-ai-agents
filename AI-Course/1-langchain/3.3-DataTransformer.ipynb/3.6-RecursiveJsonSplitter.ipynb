## 📄 JSON Splitter

- 🧩 Splits **JSON** into smaller chunks.
- 🌳 Uses **depth-first traversal**.
- 🧱 Keeps **nested objects together** when possible.
- 📏 Chunk size is based on **characters** (`min_chunk_size` & `max_chunk_size`).
- ⚠️ **Large strings are not split**.
- 💡 Use **RecursiveCharacterTextSplitter** to split large strings.
- How the text is split: --> **json value**
- How the chunk size is measured --> **By number of characters**

```python
import json
import requests

json_data = requests.get("https://api.smith.langchain.com/openapi.json").json()
json_data
```

```python
from langchain_text_splitters import RecursiveJsonSplitter

json_splitter = RecursiveJsonSplitter(max_chunk_size=1000)
json_splits = json_splitter.split_json(json_data)
json_splits
```

```python
for chunks in json_splits[:3]:
    print(chunks)
```

- Splitter can also output **documents**

```python
docs=json_splitter.create_documents([json_data])
for doc in docs[:3]:
    print(doc)
```

```python
texts=json_splitter.split_text(json_data)
print(texts[0])
print(texts[1])
```
