## How to Split by HTML Header

`HTMLHeaderTextSplitter` is a **structure-aware** text splitter that divides text based on **HTML header elements** (`<h1>`, `<h2>`, `<h3>`, etc.) and attaches metadata for each relevant header to every chunk.

### Key Features

- Splits text at the **HTML element level**.
- Adds **header metadata** to each chunk.
- Can return chunks:
  - **Element by element**, or
  - **Combined** if they share the same metadata.
- Helps in:
  - Keeping related content together.
  - Preserving the document's hierarchical structure.
  - Maintaining context-rich information.
- Can be used with **other text splitters** as part of a chunking pipeline.

### Why Use `HTMLHeaderTextSplitter`?

- Maintains the semantic structure of HTML documents.
- Improves retrieval quality for RAG applications.
- Keeps sections under the appropriate headings.
- Makes downstream processing more context-aware.

### Example

```html
<h1>Introduction</h1>
<p>This is the introduction.</p>

<h2>Features</h2>
<p>Feature 1</p>
<p>Feature 2</p>
```

The splitter will create chunks based on the HTML headers while preserving the header information as metadata.

````python
from langchain_text_splitters import HTMLHeaderTextSplitter

html_string = """
<!DOCTYPE html>
<html>
<head>
    <title>HTML Demo</title>
</head>
<body>

    <header>
        <h1>HTML Basics</h1>
        <p>Learn the most common HTML tags.</p>
    </header>

    <nav>
        <a href="#">Home</a> | <a href="#">About</a>
    </nav>

    <section>
        <h2>Text Formatting</h2>
        <p>This is <strong>bold</strong>, <em>italic</em>, and <mark>highlighted</mark> text.</p>
        <p>H<sub>2</sub>O and x<sup>2</sup></p>
    </section>

    <section>
        <h2>Lists</h2>
        <ul>
            <li>Python</li>
            <li>Java</li>
        </ul>

        <h3>Steps</h3>
        <ol>
            <li>Learn HTML</li>
            <li>Practice</li>
        </ol>
    </section>

    <section>
        <h2>Table & Form</h2>

        <table border="1">
            <tr><th>Name</th><th>Age</th></tr>
            <tr><td>Robin</td><td>24</td></tr>
        </table>

        <form>
            <label>Name:</label>
            <input type="text">
            <button>Submit</button>
        </form>
    </section>

    <footer>
        <p>&copy; 2026 HTML Demo</p>
    </footer>

</body>
</html>
"""

headers_to_split_on = [
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3")
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
splits = html_splitter.split_text(html_string)
splits
````

````python
import requests
from langchain_text_splitters import HTMLHeaderTextSplitter

url = "https://en.wikipedia.org/wiki/HTML"

headers = {
    "User-Agent": "Mozilla/5.0"
}

html = requests.get(url, headers=headers).text

html_splitter = HTMLHeaderTextSplitter([
    ("h1", "Header 1"),
    ("h2", "Header 2"),
    ("h3", "Header 3"),
])

splits = html_splitter.split_text(html)
splits[:3]  # Display the first 3 splits
````
